```python
import pandas as pd
from pycaret.classification import *

import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the first row to get the column headers
columns = pd.read_csv("../data/source/X_train.csv", nrows=1)

# Read all columns except for the MDR_text column
df_X_train = pd.read_csv("../data/source/X_train.csv", usecols=[x for x in columns if x != "MDR_text"])

df_y_train = pd.read_csv("../data/source/y_train.csv")

df_train = pd.concat([df_y_train, df_X_train], axis=1)

print(df_train.shape)
df_train.head(1).T
```

```python
setup(data = df_train, target='recall', session_id=365, silent=True, fold=5)
#      log_experiment=True, experiment_name="02_train_no_text")
```
