### 1.2.2 Load normalized texts

```python
df_normalized = pd.read_csv("../data/prepared/train_normalized.csv")

print(df_normalized.shape)
df_normalized[["MDR_text","normalized"]].head()
```

### 1.2.3 - Vectorize the texts (unfiltered)

- min_df = 0.0
- max_df = 1.0

```python
df_unfiltered = preprocess.vectorize(df_normalized, "normalized", "tfidf", 1, 1, 0.0,1.0)

print(df_unfiltered.shape)

df_unfiltered.sample(5)
```

```python
df_unfiltered.to_csv("../data/prepared/train_vectorized_unfiltered.csv", index=False)
```

### 1.2.4 - Vectorize the texts (filtered)

Filtered to reduce the number of features.

- min_df = 0.05 
  - remove tokens that appear in less than 5% of texts (rare words)
- max_df = 0.75
  - remove tokens that appear in more than 75% of texts (common words)

```python
df_filtered = preprocess.vectorize(df_normalized, "normalized", "tfidf", 1, 1, 0.05, 0.70)

print(df_filtered.shape)

df_filtered.sample(5)
```

```python
df_filtered.to_csv("../data/prepared/train_vectorized_filtered.csv", index=False)
```
