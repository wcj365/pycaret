```python
df_unfiltered = pd.read_csv("../data/prepared/train_vectorized_unfiltered.csv")

print(df_unfiltered.shape)

df_unfiltered.head()
```

```python
df_y_train = pd.read_csv("../data/source/y_train.csv")
df_y_train.head()
```

```python
# To avoid the possibility "recall" is one of the features

df_y_train.rename(columns={"recall":"RECALL"}, inplace=True)
```

```python
df_combine = pd.concat([df_y_train, df_unfiltered], axis=1)
df_combine.head()
```

```python
df_chi_squared = preprocess.chi_squared(df_combine, "RECALL", 0.5)

print(df_chi_squared.shape)

df_chi_squared.head(1).T
```

```python
df_chi_squared.head()
```

```python
df_chi_squared.drop(columns=["RECALL"]).to_csv("../data/prepared/train_chi_squared_p50.csv", index=False)
```
