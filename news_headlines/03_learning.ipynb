```python
import pandas as pd
from pycaret.classification import *

import warnings
warnings.filterwarnings('ignore')
```

```python
df_learn = pd.read_csv("data/preprocessed.csv")
print(df_learn.shape)
df_learn.head()
```

```python
setup(data = df_learn, 
              target = 'category', 
              silent=True,
              ignore_features=['headline','Dominant_Topic','Perc_Dominant_Topic'],
              log_experiment=True,
              experiment_name="02_learning")
```

```python
best = compare_models()
```

```python
model_rf = create_model("rf")
```

```python
tuned_rf = tune_model(model_rf)
```

```python
evaluate_model(tuned_rf)
```

```python
# Use the hold-out observations for testing the model

predict_model(tuned_rf)
```

```python
# Finalize the model

final_rf = finalize_model(tuned_rf)
print(final_rf)
```

```python
save_model(final_rf,'models/random_forest')
```
