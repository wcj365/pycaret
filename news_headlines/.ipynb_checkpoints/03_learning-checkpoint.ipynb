```python
import pandas as pd
from pycaret.nlp import *

import warnings
warnings.filterwarnings('ignore')
```

```python
df_learn = pd.read_csv("data/news_headlines.csv")
print(df_learn.shape)
df_learn.head()
```

```python
s = setup(data=df_learn.sample(500).reset_index(drop=True), target="headline")
 #   log_experiment=True, experiment_name="random_forest")
```

```python
lda = create_model('lda', multi_core =True)
```

```python
lda_data = assign_model(lda)
lda_data.head()
```

```python
lda_data.shape
```

```python
evaluate_model(lda)
```

```python
from pycaret.classification import *
```

```python
model = setup(data = lda_data, target = 'category', silent=True,
              ignore_features=['headline','Dominant_Topic','Perc_Dominant_Topic'])
```

```python
best = compare_models()
```

```python
model_rf = create_model("rf")
```

```python
tuned_rf = tune_model(model_rf)
```

```python
evaluate_model(tuned_rf)
```

```python
# Use the hold-out observations for testing the model

predict_model(tuned_rf)
```

```python
# Finalize the model

final_rf = finalize_model(tuned_rf)
print(final_rf)
```

```python
save_model(final_rf,'models/random_forest')
```
